```python
train_df, temp_df = train_test_split(
    labels_df,
    test_size  = CONFIG['VAL_SPLIT'] + CONFIG['TEST_SPLIT'],
    stratify   = labels_df['binary_label'],
    random_state = CONFIG['RANDOM_SEED']
)
val_df, test_df = train_test_split(
    temp_df,
    test_size  = CONFIG['TEST_SPLIT'] / (CONFIG['VAL_SPLIT'] + CONFIG['TEST_SPLIT']),
    stratify   = temp_df['binary_label'],
    random_state = CONFIG['RANDOM_SEED']
)

print(f'Train : {len(train_df):,}  '
      f'(benign={(train_df.binary_label==0).sum():,}  '
      f'malignant={(train_df.binary_label==1).sum():,})')
print(f'Val   : {len(val_df):,}')
print(f'Test  : {len(test_df):,}')

cw = compute_class_weight(
    class_weight = 'balanced',
    classes      = np.array([0, 1]),
    y            = train_df['binary_label'].values
)
CLASS_WEIGHTS = {0: float(cw[0]), 1: float(cw[1])}
print(f'\nClass weights: {CLASS_WEIGHTS}')
```

```python
# ── Version-safe augmentation pipeline ───────────────────────────
if ALB_VER >= (1, 4):
    _coarse = A.CoarseDropout(
        num_holes_upper=8, hole_height_upper=32,
        hole_width_upper=32, fill_value=0, p=0.3
    )
else:
    _coarse = A.CoarseDropout(
        max_holes=8, max_height=32, max_width=32,
        fill_value=0, p=0.3
    )

if ALB_VER >= (2, 0):
    _noise = A.GaussNoise(std_range=(0.03, 0.2), p=1.0)
else:
    _noise = A.GaussNoise(var_limit=(10.0, 50.0), p=1.0)

TRAIN_AUG = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.3),
    A.RandomRotate90(p=0.4),
    A.ShiftScaleRotate(
        shift_limit=0.1, scale_limit=0.15,
        rotate_limit=45, p=0.4
    ),
    A.OneOf([
        _noise,
        A.GaussianBlur(blur_limit=3, p=1.0),
        A.MedianBlur(blur_limit=3, p=1.0),
    ], p=0.3),
    A.OneOf([
        A.CLAHE(clip_limit=2.0, p=1.0),
        A.RandomBrightnessContrast(
            brightness_limit=0.2, contrast_limit=0.2, p=1.0
        ),
    ], p=0.4),
    A.HueSaturationValue(
        hue_shift_limit=10, sat_shift_limit=20,
        val_shift_limit=10, p=0.3
    ),
    _coarse,
])

print('Augmentation pipeline ready.')
```

```python
def load_image(path, label, augment=False):
    """
    Read JPEG → resize → optional augment → EfficientNet preprocess.
    Label is kept as int32 (0 or 1) — matches sigmoid binary output.
    """
    raw = tf.io.read_file(path)
    img = tf.image.decode_jpeg(raw, channels=3)
    img = tf.image.resize(img, [IMG_SIZE, IMG_SIZE])
    img = tf.cast(img, tf.uint8)

    if augment:
        def _aug(x):
            return TRAIN_AUG(image=x.numpy())['image'].astype(np.uint8)
        img = tf.py_function(_aug, [img], tf.uint8)
        img.set_shape([IMG_SIZE, IMG_SIZE, 3])

    img = tf.cast(img, tf.float32)
    img = preprocess_input(img)           # maps [0,255] to [-1, 1]
    # Cast label to float32 for binary cross-entropy with sigmoid output
    return img, tf.cast(label, tf.float32)


def build_dataset(df, augment=False, shuffle=False):
    """
    Build a batched tf.data.Dataset.

    KEY FIX: drop_remainder=True ensures every batch is exactly
    BATCH_SIZE.  Without this, the last smaller batch causes
    'Incompatible shapes' crashes inside BatchNorm / XLA.
    """
    paths  = [os.path.join(CONFIG['IMAGE_DIR'], f'{n}.jpg')
              for n in df['image_name']]
    labels = df['binary_label'].values.astype(np.int32)

    pairs = [(p, l) for p, l in zip(paths, labels)
             if os.path.exists(p)]

    if not pairs:
        # Synthetic fallback so notebook runs without real images
        print('WARNING: no images found — using synthetic data.')
        n = (min(len(df), 512) // CONFIG['BATCH_SIZE']) * CONFIG['BATCH_SIZE']
        imgs = tf.random.uniform((n, IMG_SIZE, IMG_SIZE, 3), dtype=tf.float32)
        lbls = tf.cast(tf.constant(labels[:n]), tf.float32)
        ds   = tf.data.Dataset.from_tensor_slices((imgs, lbls))
        if shuffle: ds = ds.shuffle(n, seed=CONFIG['RANDOM_SEED'])
        return (ds.batch(CONFIG['BATCH_SIZE'], drop_remainder=True)
                  .prefetch(tf.data.AUTOTUNE))

    vp, vl = zip(*pairs)
    ds = tf.data.Dataset.from_tensor_slices((list(vp), list(vl)))
    if shuffle:
        ds = ds.shuffle(len(vp), seed=CONFIG['RANDOM_SEED'],
                        reshuffle_each_iteration=True)
    ds = ds.map(
        lambda p, l: load_image(p, l, augment=augment),
        num_parallel_calls=tf.data.AUTOTUNE
    )
    return (ds.batch(CONFIG['BATCH_SIZE'], drop_remainder=True)
              .prefetch(tf.data.AUTOTUNE))


train_ds = build_dataset(train_df, augment=True,  shuffle=True)
val_ds   = build_dataset(val_df,   augment=False, shuffle=False)
test_ds  = build_dataset(test_df,  augment=False, shuffle=False)

print(f'Train batches : {len(train_ds)}')
print(f'Val batches   : {len(val_ds)}')
print(f'Test batches  : {len(test_ds)}')
```

```python
def denorm(t):
    """Reverse EfficientNet preprocess_input for display."""
    x = t.numpy() if hasattr(t, 'numpy') else np.array(t)
    return np.clip((x + 1.0) / 2.0, 0, 1)

for imgs, lbls in train_ds.take(1):
    n_show = min(18, len(imgs))
    fig, axes = plt.subplots(3, 6, figsize=(16, 8))
    fig.suptitle('Sample Training Images', fontsize=12, fontweight='bold')
    for i, ax in enumerate(axes.flat):
        if i >= n_show: ax.axis('off'); continue
        ax.imshow(denorm(imgs[i]))
        lbl_idx = int(round(float(lbls[i])))
        ax.set_title(CONFIG['BINARY_NAMES'][lbl_idx], fontsize=7,
                     color='#F44336' if lbl_idx == 1 else '#4CAF50')
        ax.axis('off')
    plt.tight_layout()
    plt.savefig(f"{CONFIG['RESULTS_DIR']}/sample_images.png", bbox_inches='tight')
    plt.show()
```

```python
def build_model(img_size=224, dropout=0.4, freeze_base=True):
    inputs = Input(shape=(img_size, img_size, 3), name='image_input')

    base = EfficientNetB3(
        include_top=False,
        weights='imagenet',
        input_tensor=inputs
    )
    base.trainable = not freeze_base

    x   = base.output
    x   = layers.GlobalAveragePooling2D(name='gap')(x)
    x   = layers.BatchNormalization(name='bn1')(x)
    x   = layers.Dropout(dropout, name='drop1')(x)
    x   = layers.Dense(
              512, activation='relu', name='fc1',
              kernel_regularizer=keras.regularizers.l2(1e-4)
          )(x)
    x   = layers.BatchNormalization(name='bn2')(x)
    x   = layers.Dropout(dropout / 2, name='drop2')(x)
    # SINGLE neuron, sigmoid — produces a scalar probability per sample
    out = layers.Dense(1, activation='sigmoid', name='output')(x)

    return Model(inputs, out, name='DermScan_EfficientNetB3'), base


model, base_model = build_model(
    img_size    = IMG_SIZE,
    dropout     = CONFIG['DROPOUT'],
    freeze_base = True
)

total     = model.count_params()
trainable = sum(tf.size(w).numpy() for w in model.trainable_weights)
print(f'Total params     : {total:,}')
print(f'Trainable params : {trainable:,}')
print(f'Output shape     : {model.output_shape}   <- must be (None, 1)')
```

```python
focal_loss = keras.losses.BinaryFocalCrossentropy(
    gamma              = 2.0,
    apply_class_balancing = True,   # handles benign-heavy imbalance
    from_logits        = False
)


def compile_model(mdl, lr):
    """
    All metrics here expect a (batch,) or (batch, 1) sigmoid output
    and float32 labels — which is exactly what our pipeline produces.
    No shape mismatches possible.
    """
    mdl.compile(
        optimizer = Adam(learning_rate=lr, clipnorm=1.0),
        loss      = focal_loss,
        metrics   = [
            keras.metrics.BinaryAccuracy(name='accuracy', threshold=0.5),
            keras.metrics.AUC(name='auc', from_logits=False),
            keras.metrics.Precision(name='precision', thresholds=0.5),
            keras.metrics.Recall(name='recall',    thresholds=0.5),
        ]
    )


compile_model(model, CONFIG['LR_STAGE1'])
print('Model compiled — Stage 1 (frozen backbone).')
```

```python
def make_callbacks(stage):
    return [
        ModelCheckpoint(
            filepath     = CONFIG['MODEL_SAVE_PATH'],
            monitor      = 'val_auc',
            mode         = 'max',
            save_best_only = True,
            verbose      = 1
        ),
        EarlyStopping(
            monitor              = 'val_auc',
            mode                 = 'max',
            patience             = CONFIG['EARLY_STOP_PAT'],
            restore_best_weights = True,
            verbose              = 1
        ),
        ReduceLROnPlateau(
            monitor  = 'val_loss',
            factor   = 0.5,
            patience = 3,
            min_lr   = 1e-8,
            verbose  = 1
        ),
        TensorBoard(
            log_dir        = f"{CONFIG['LOG_DIR']}/stage{stage}",
            histogram_freq = 0
        ),
        CSVLogger(
            f"{CONFIG['RESULTS_DIR']}/history_stage{stage}.csv"
        ),
    ]
```

```python
print('=' * 60)
print('STAGE 1 — Train head only (backbone frozen)')
print(f'Epochs : {CONFIG["STAGE1_EPOCHS"]}  |  LR : {CONFIG["LR_STAGE1"]}')
print('=' * 60)

hist1 = model.fit(
    train_ds,
    validation_data = val_ds,
    epochs          = CONFIG['STAGE1_EPOCHS'],
    class_weight    = CLASS_WEIGHTS,
    callbacks       = make_callbacks(1),
    verbose         = 1
)
print('Stage 1 complete.')
```

```python
print('=' * 60)
print('STAGE 2 — Fine-tune top backbone layers')
print(f'Unfreeze from layer {CONFIG["FINE_TUNE_FROM"]}  |  LR : {CONFIG["LR_STAGE2"]}')
print('=' * 60)

base_model.trainable = True
for layer in base_model.layers[:CONFIG['FINE_TUNE_FROM']]:
    layer.trainable = False

print(f'Trainable backbone layers : '
      f'{sum(1 for l in base_model.layers if l.trainable)}')

compile_model(model, CONFIG['LR_STAGE2'])

hist2 = model.fit(
    train_ds,
    validation_data = val_ds,
    epochs          = CONFIG['STAGE1_EPOCHS'] + CONFIG['STAGE2_EPOCHS'],
    initial_epoch   = len(hist1.history['loss']),
    class_weight    = CLASS_WEIGHTS,
    callbacks       = make_callbacks(2),
    verbose         = 1
)
print('Stage 2 complete.')
```

```python
res = model.evaluate(test_ds, verbose=1)
print('\n' + '='*55 + '\nTEST SET RESULTS\n' + '='*55)
for k, v in zip(['loss','accuracy','auc','precision','recall'], res):
    print(f'  {k:12s}: {v:.4f}')
print('='*55)
```

```python
all_true, all_pred, all_prob = [], [], []

for imgs, lbls in tqdm(test_ds, desc='Predicting'):
    # model outputs shape (batch, 1) — squeeze to (batch,)
    probs = tf.squeeze(model(imgs, training=False), axis=-1).numpy()
    preds = (probs >= 0.5).astype(int)
    all_true.extend(lbls.numpy().astype(int))
    all_pred.extend(preds)
    all_prob.extend(probs)

all_true = np.array(all_true)
all_pred = np.array(all_pred)
all_prob = np.array(all_prob)

try:
    auc_val = roc_auc_score(all_true, all_prob)
except ValueError:
    auc_val = 0.0

print(f'ROC-AUC  : {auc_val:.4f}')
print(f'F1 Macro : {sk_f1(all_true, all_pred, average="macro", zero_division=0):.4f}')
print('\nClassification Report:')
print(classification_report(
    all_true, all_pred,
    target_names=CONFIG['BINARY_NAMES'], zero_division=0
))
```

```python
def find_gradcam_layer(mdl, preferred='top_conv'):
    """Return preferred layer or last Conv2D as fallback."""
    try:
        mdl.get_layer(preferred); return preferred
    except ValueError:
        pass
    for lyr in reversed(mdl.layers):
        if isinstance(lyr, layers.Conv2D):
            print(f'Grad-CAM fallback layer: {lyr.name}')
            return lyr.name
    raise ValueError('No Conv2D layer found.')


class GradCAM:
    """Grad-CAM using tf.GradientTape. Works with sigmoid output."""

    def __init__(self, mdl, layer_name=None):
        if layer_name is None:
            layer_name = find_gradcam_layer(mdl)
        print(f'Grad-CAM target layer: {layer_name}')
        self.grad_model = Model(
            inputs  = mdl.inputs,
            outputs = [mdl.get_layer(layer_name).output, mdl.output]
        )

    def compute(self, img_batch):
        """
        img_batch : preprocessed float32, shape (1, H, W, 3)
        Returns   : heatmap (H, W), pred_class (0/1), confidence (float)
        """
        img_batch = tf.cast(img_batch, tf.float32)
        with tf.GradientTape() as tape:
            tape.watch(img_batch)
            conv_out, pred = self.grad_model(img_batch, training=False)
            # pred shape: (1, 1) for sigmoid output
            score = pred

        grads   = tape.gradient(score, conv_out)   # (1, H, W, C)
        weights = tf.reduce_mean(grads, axis=(0, 1, 2))  # (C,)
        cam     = tf.reduce_sum(conv_out[0] * weights, axis=-1)  # (H, W)
        cam     = tf.nn.relu(cam).numpy()
        if cam.max() > 1e-8:
            cam = cam / cam.max()

        prob       = float(tf.squeeze(pred))
        pred_class = int(prob >= 0.5)
        return cam, pred_class, prob

    def overlay(self, img_uint8, heatmap, alpha=0.45):
        h, w = img_uint8.shape[:2]
        hm   = cv2.resize(heatmap, (w, h))
        hm_c = cv2.applyColorMap((hm * 255).astype(np.uint8),
                                  cv2.COLORMAP_JET)
        hm_c = cv2.cvtColor(hm_c, cv2.COLOR_BGR2RGB)
        return cv2.addWeighted(img_uint8, 1 - alpha, hm_c, alpha, 0)


gradcam = GradCAM(model)
print('GradCAM ready.')
```

```python
class LesionChangeAnalyzer:
    """
    Compares two images of the same lesion and returns a risk report.
    Metrics: SSIM, area, compactness, color variance, asymmetry.
    """

    def __init__(self, size=(224, 224),
                 threshold=CONFIG['CHANGE_THRESHOLD']):
        self.size      = size
        self.threshold = threshold

    def _load(self, src):
        if isinstance(src, str):
            return np.array(Image.open(src).convert('RGB').resize(self.size))
        if isinstance(src, Image.Image):
            return np.array(src.convert('RGB').resize(self.size))
        arr = np.array(src, dtype=np.uint8)
        return cv2.resize(arr, self.size) if arr.shape[:2] != self.size else arr

    def _segment(self, rgb):
        gray = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)
        blur = cv2.GaussianBlur(gray, (7, 7), 0)
        _, m = cv2.threshold(
            blur, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
        k = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (7, 7))
        m = cv2.morphologyEx(m, cv2.MORPH_CLOSE, k, iterations=2)
        m = cv2.morphologyEx(m, cv2.MORPH_OPEN,  k, iterations=1)
        return m

    def _features(self, rgb):
        mask = self._segment(rgb)
        area = float(np.sum(mask > 0)) / float(mask.size)

        cnts, _ = cv2.findContours(
            mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        if cnts:
            c    = max(cnts, key=cv2.contourArea)
            peri = float(cv2.arcLength(c, True))
            comp = (peri**2) / (4*np.pi*(float(cv2.contourArea(c))+1e-6))
        else:
            peri = comp = 0.0

        hsv  = cv2.cvtColor(rgb, cv2.COLOR_RGB2HSV).astype(float)
        px   = hsv[mask > 0] if mask.sum() > 0 else hsv.reshape(-1,3)
        cvar = float(np.mean(np.std(px, axis=0)))

        h, w  = mask.shape
        t, b  = float(mask[:h//2].sum()), float(mask[h//2:].sum())
        asym  = abs(t - b) / (t + b + 1e-6)

        return dict(area=area, perimeter=peri, compactness=float(comp),
                    color_var=cvar, asymmetry=float(asym), mask=mask)

    def analyze(self, before, after,
                date_before='Visit 1', date_after='Visit 2'):
        b, a   = self._load(before), self._load(after)
        fb, fa = self._features(b),  self._features(a)

        sv = ssim(
            cv2.cvtColor(b, cv2.COLOR_RGB2GRAY),
            cv2.cvtColor(a, cv2.COLOR_RGB2GRAY),
            data_range=255
        )
        d = {
            'area_delta'       : fa['area']        - fb['area'],
            'compactness_delta': fa['compactness'] - fb['compactness'],
            'color_var_delta'  : fa['color_var']   - fb['color_var'],
            'asymmetry_delta'  : fa['asymmetry']   - fb['asymmetry'],
            'ssim'             : float(sv),
            'ssim_change'      : float(1.0 - sv),
        }
        risk = (
            abs(d['area_delta'])        * 2.0 +
            abs(d['compactness_delta']) * 1.5 +
            abs(d['color_var_delta'])   * 0.5 / 20 +
            abs(d['asymmetry_delta'])   * 1.0 +
                d['ssim_change']        * 1.0
        ) / 6.0

        level = ('HIGH'   if risk > 0.3 else
                 'MEDIUM' if risk > self.threshold else 'LOW')
        return dict(
            date_before=date_before, date_after=date_after,
            before_feats=fb, after_feats=fa, deltas=d,
            risk_score=float(risk), risk_level=level,
            alert=risk > self.threshold,
            img_before=b, img_after=a
        )


analyzer = LesionChangeAnalyzer()
print('LesionChangeAnalyzer ready.')
```

```python
def plot_change_report(rpt):
    CLR = {'HIGH':'#F44336','MEDIUM':'#FF9800','LOW':'#4CAF50'}
    fig = plt.figure(figsize=(18, 8))
    fig.suptitle(
        f'Lesion Change  {rpt["date_before"]} → {rpt["date_after"]}'
        f'   Risk: {rpt["risk_level"]}  ({rpt["risk_score"]:.3f})',
        fontsize=12, fontweight='bold', color=CLR[rpt['risk_level']]
    )
    gs = gridspec.GridSpec(2, 5, hspace=0.45, wspace=0.35)

    diff = cv2.cvtColor(
        cv2.absdiff(rpt['img_before'], rpt['img_after']),
        cv2.COLOR_RGB2GRAY)
    for idx,(img,ttl,cmap) in enumerate([
        (rpt['img_before'],            'Before',      None),
        (rpt['img_after'],             'After',       None),
        (diff,  f'Diff\nSSIM={rpt["deltas"]["ssim"]:.3f}', 'gray'),
        (rpt['before_feats']['mask'],  'Mask Before', 'gray'),
        (rpt['after_feats']['mask'],   'Mask After',  'gray'),
    ]):
        ax = fig.add_subplot(gs[0, idx])
        ax.imshow(img, cmap=cmap); ax.set_title(ttl, fontsize=9); ax.axis('off')

    ax6 = fig.add_subplot(gs[1, :3])
    d   = rpt['deltas']
    keys= ['Area','Compact.','Color/20','Asymmetry','SSIM chg']
    vals= [d['area_delta'], d['compactness_delta'],
           d['color_var_delta']/20, d['asymmetry_delta'], d['ssim_change']]
    bc  = ['#F44336' if v > 0 else '#4CAF50' for v in vals]
    bars= ax6.barh(keys, vals, color=bc, edgecolor='white')
    ax6.axvline(0, color='black', lw=0.8)
    ax6.set_title('Feature Deltas (red = increase = concerning)', fontsize=9)
    for bar, v in zip(bars, vals):
        ax6.text(v+0.001, bar.get_y()+bar.get_height()/2,
                 f'{v:+.4f}', va='center', fontsize=8)

    ax7  = fig.add_subplot(gs[1, 3:])
    risk = rpt['risk_score']
    ax7.barh(['Risk'], [1.0], color='#EEEEEE', edgecolor='grey')
    ax7.barh(['Risk'], [min(risk,1.0)], color=CLR[rpt['risk_level']])
    ax7.set_xlim(0, 1)
    ax7.axvline(CONFIG['CHANGE_THRESHOLD'], color='orange',
                ls='--', lw=1.5, label='Alert threshold')
    ax7.axvline(0.3, color='red', ls='--', lw=1.5, label='High risk')
    ax7.text(min(risk,0.9)+0.02, 0, f'{risk:.3f}',
             va='center', fontsize=11, fontweight='bold',
             color=CLR[rpt['risk_level']])
    ax7.legend(fontsize=8, loc='lower right')
    ax7.set_title(
        f'Risk  Alert:{" YES" if rpt["alert"] else " No"} '
        f' Level:{rpt["risk_level"]}', fontsize=9)

    plt.savefig(f"{CONFIG['RESULTS_DIR']}/growth_report.png",
                bbox_inches='tight')
    plt.show()


# Demo using two test-set images as before/after
for imgs, _ in test_ds.take(1):
    v1_np = (denorm(imgs[0]) * 255).astype(np.uint8)
    v2_np = (denorm(imgs[1]) * 255).astype(np.uint8)

v2_mod = cv2.convertScaleAbs(v2_np, alpha=1.06, beta=6)
v2_mod = cv2.GaussianBlur(v2_mod, (3,3), 0)

report = analyzer.analyze(v1_np, v2_mod,
                           date_before='2024-02-10',
                           date_after='2024-08-22')
plot_change_report(report)
print(f"Risk  : {report['risk_level']} ({report['risk_score']:.4f})")
print(f"Alert : {'Consult a dermatologist' if report['alert'] else 'Stable'}")
```

```python
class PatientLesionTracker:
    """Multi-visit per-patient lesion timeline with CNN prediction."""

    def __init__(self):
        self.records  = {}
        self.analyzer = LesionChangeAnalyzer()

    def add_visit(self, pid, image_np, date=None, mdl=None):
        if date is None:
            date = datetime.now().strftime('%Y-%m-%d')
        image_np = np.array(image_np, dtype=np.uint8)

        pred_lbl, conf = 'N/A', 0.0
        if mdl is not None:
            t = preprocess_input(
                tf.cast(
                    tf.image.resize(image_np, [IMG_SIZE, IMG_SIZE]),
                    tf.float32
                )
            )[tf.newaxis]
            prob = float(tf.squeeze(mdl(t, training=False)))
            pred_lbl = CONFIG['BINARY_NAMES'][int(prob >= 0.5)]
            conf     = prob if prob >= 0.5 else 1 - prob

        rec = dict(date=date, image=image_np,
                   prediction=pred_lbl, confidence=conf,
                   change_report=None)

        if self.records.get(pid):
            prev = self.records[pid][-1]
            rec['change_report'] = self.analyzer.analyze(
                prev['image'], image_np,
                date_before=prev['date'], date_after=date
            )
        self.records.setdefault(pid, []).append(rec)
        return rec

    def plot_timeline(self, pid):
        visits = self.records.get(pid, [])
        if not visits: print('No records.'); return
        n = len(visits)
        fig, axes = plt.subplots(2, n, figsize=(4*n, 6))
        if n == 1: axes = np.array(axes).reshape(2,1)
        fig.suptitle(f'Patient {pid} — Lesion Timeline',
                     fontsize=12, fontweight='bold')
        CLR = {'HIGH':'#F44336','MEDIUM':'#FF9800','LOW':'#4CAF50'}
        for i, v in enumerate(visits):
            cr    = v['change_report']
            risk  = cr['risk_score'] if cr else 0.0
            level = cr['risk_level'] if cr else 'LOW'
            axes[0,i].imshow(v['image'])
            axes[0,i].set_title(
                f"{v['date']}\n{v['prediction']}\n({v['confidence']:.0%})",
                fontsize=8)
            axes[0,i].axis('off')
            axes[1,i].bar(['Risk'], [max(risk,0.01)], color=CLR[level])
            axes[1,i].set_ylim(0, 0.6)
            axes[1,i].set_title(f'Risk:{risk:.3f}\n{level}', fontsize=8)
            axes[1,i].axhline(CONFIG['CHANGE_THRESHOLD'],
                               color='orange', ls='--', lw=1)
        plt.tight_layout()
        plt.savefig(
            f"{CONFIG['RESULTS_DIR']}/patient_{pid}_timeline.png",
            bbox_inches='tight')
        plt.show()


tracker = PatientLesionTracker()
visit1 = v1_np.copy()
visit2 = cv2.convertScaleAbs(v1_np, alpha=1.04, beta=4)
visit3 = cv2.convertScaleAbs(v1_np, alpha=1.13, beta=12)

for img, dt in [(visit1,'2024-01-10'),(visit2,'2024-05-18'),(visit3,'2024-09-25')]:
    tracker.add_visit('PAT-001', img, date=dt, mdl=model)

tracker.plot_timeline('PAT-001')
```

```python
def predict(image_input, mdl,
             gradcam_obj=None, anlzr=None, prior_image=None):
    """Full inference. Accepts file path, PIL Image, or numpy array."""
    if isinstance(image_input, str):
        pil = Image.open(image_input).convert('RGB')
    elif isinstance(image_input, Image.Image):
        pil = image_input.convert('RGB')
    else:
        pil = Image.fromarray(np.array(image_input, dtype=np.uint8))

    img_np  = np.array(pil.resize((IMG_SIZE, IMG_SIZE)))
    img_pre = preprocess_input(img_np.astype(np.float32))
    batch   = img_pre[np.newaxis]

    prob       = float(tf.squeeze(mdl(batch, training=False)))
    pred_class = int(prob >= 0.5)
    conf       = prob if pred_class == 1 else 1 - prob

    cam_overlay = None
    if gradcam_obj is not None:
        try:
            hm, _, _ = gradcam_obj.compute(tf.constant(batch))
            cam_overlay = gradcam_obj.overlay(img_np, hm)
        except Exception as e:
            print(f'Grad-CAM warning: {e}')

    change_rep = None
    if anlzr is not None and prior_image is not None:
        prior_np = (
            np.array(prior_image.convert('RGB'))
            if isinstance(prior_image, Image.Image)
            else np.array(prior_image, dtype=np.uint8)
        )
        change_rep = anlzr.analyze(prior_np, img_np)

    return {
        'predicted_class' : CONFIG['BINARY_NAMES'][pred_class],
        'class_index'     : pred_class,
        'confidence'      : conf,
        'prob_malignant'  : prob,
        'gradcam_overlay' : cam_overlay,
        'original_image'  : img_np,
        'change_report'   : change_rep,
    }


def display_result(res):
    has_cam = res['gradcam_overlay'] is not None
    has_cr  = res['change_report']   is not None
    ncols   = 2 + int(has_cam) + int(has_cr)
    fig, axes = plt.subplots(1, ncols, figsize=(5*ncols, 5))
    c = '#F44336' if res['class_index'] == 1 else '#4CAF50'
    fig.suptitle(
        f'Derm-Scan: {res["predicted_class"]}  '
        f'(P(malignant)={res["prob_malignant"]:.1%})',
        fontsize=12, fontweight='bold', color=c
    )

    ai = 0
    axes[ai].imshow(res['original_image'])
    axes[ai].set_title('Input Image', fontsize=10)
    axes[ai].axis('off'); ai += 1

    if has_cam:
        axes[ai].imshow(res['gradcam_overlay'])
        axes[ai].set_title('Grad-CAM', fontsize=10)
        axes[ai].axis('off'); ai += 1

    p = res['prob_malignant']
    axes[ai].barh(
        CONFIG['BINARY_NAMES'],
        [1-p, p],
        color=['#4CAF50','#F44336']
    )
    axes[ai].set_xlim(0, 1)
    for j, v in enumerate([1-p, p]):
        axes[ai].text(v+0.01, j, f'{v:.1%}', va='center', fontsize=9)
    axes[ai].set_title('Probabilities', fontsize=10); ai += 1

    if has_cr:
        cr = res['change_report']
        txt = (
            f"Risk Level  : {cr['risk_level']}\n"
            f"Risk Score  : {cr['risk_score']:.4f}\n"
            f"SSIM Change : {cr['deltas']['ssim_change']:.4f}\n"
            f"Area Delta  : {cr['deltas']['area_delta']:+.4f}\n"
            f"Color Var D : {cr['deltas']['color_var_delta']:+.2f}\n"
            f"Asymmetry D : {cr['deltas']['asymmetry_delta']:+.4f}\n"
            f"Alert       : {'YES' if cr['alert'] else 'No'}"
        )
        axes[ai].text(
            0.1, 0.5, txt, transform=axes[ai].transAxes,
            fontsize=10, va='center', family='monospace',
            bbox=dict(boxstyle='round', facecolor='#FFF3E0', alpha=0.85)
        )
        axes[ai].set_title('Growth Tracking', fontsize=10)
        axes[ai].axis('off')

    plt.tight_layout()
    plt.savefig(f"{CONFIG['RESULTS_DIR']}/inference_result.png",
                bbox_inches='tight')
    plt.show()


result = predict(
    visit1, model,
    gradcam_obj = gradcam,
    anlzr       = analyzer,
    prior_image = visit2
)
display_result(result)
print(f"Prediction : {result['predicted_class']}")
print(f"P(malig.)  : {result['prob_malignant']:.1%}")
if result['change_report']:
    print(f"Growth risk: {result['change_report']['risk_level']} "
          f"({result['change_report']['risk_score']:.4f})")
```
